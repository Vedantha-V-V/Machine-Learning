```python
import pandas as pd
df = pd.read_csv('pokemon.csv',encoding='unicode_escape')
df.head()
```

```python
types = df['Type 1'].isin(['Grass','Fire','Water'])
drop_cols = ['Type 1','Type 2','Generation','Legendary']
df = df[types].drop(columns = drop_cols)
df.head()
```

```python
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=3,random_state=3)
df['cluster'] = kmeans.fit_predict(df[['Attack','Defense']])

centroids = kmeans.cluster_centers_
cen_x = [i[0] for i in centroids]
cen_y = [i[0] for i in centroids]

df['cen_x'] = df.cluster.map({0:cen_x[0], 1:cen_x[1], 2:cen_x[2]})
df['cen_y'] = df.cluster.map({0:cen_y[0], 1:cen_y[1], 2:cen_y[2]})

colors = ['#DF2020','#B1DF20','#2095DF']
df['c'] = df.cluster.map({0:colors[0], 1:colors[1], 2:colors[2]})
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df.Attack, df.Defense, c=df.c, alpha=0.6, s=10)
```

```python
plt.scatter(df.Attack, df.Defense, c=df.c, alpha=0.6)
```

```python
#Creating Centroid point for each cluster
from matplotlib.lines import Line2D
fig, ax = plt.subplots(1, figsize = (8,8))
plt.scatter(df.Attack, df.Defense, c=df.c, alpha=0.6, s=10)
plt.scatter(cen_x, cen_y, marker='*',c=colors, s=70)
for idx, val in df.iterrows():
    x=[val.Attack, val.cen_x]
    y=[val.Defense, val.cen_y]
    plt.plot(x,y,c=val.c, alpha=0.6)
    
#Centroids
cent_leg = [Line2D([0], [1], marker='*',label='Centroid -c{}'.format(i+1),markerfacecolor=color,
                  markersize=10) for i,color in enumerate(colors)]

#Legend
legend_elements = [Line2D([0], [1], marker='*',label='Centroid -c{}'.format(i+1),markerfacecolor=color,
                  markersize=10) for i,color in enumerate(colors)]
legend_elements.extend([Line2D([0], [1], marker='*',label='Centroid -c{}'.format(i+1),markerfacecolor=color,
                  markersize=10) for i,color in enumerate(colors)])

plt.legend(handles=legend_elements, loc='upper right',ncol=2)
plt.xlim(0,200)
plt.ylim(0,200)
plt.title('Pokemon Cluster', loc ='left',fontsize=22)
plt.xlabel('Attack')
plt.ylabel('Defense')
```
